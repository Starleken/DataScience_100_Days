```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
np.random.seed(0)
X = np.random.rand(100, 1) 
y = 5 * X + 1000 + 0.1 * np.random.randn(100, 1) 
iteration_count = 1000
bias = 0
slope = 0
learning_rate = 0.5

len_data = len(X)
error = 0
for i in range (iteration_count):
    predict = bias +  slope * X

    error = np.sum(y - predict)
    
    slope_gradient = -X.T.dot(y - predict)/len_data
  
    bias_gradient = -np.sum(y - predict)/len_data

    slope = slope - learning_rate * slope_gradient
    bias = bias - learning_rate * bias_gradient

plt.scatter(X, y, label='Исходные данные')
plt.plot(X, X * slope + bias, color='red', linestyle='--', label='Линейная регрессия (обученная)')
plt.xlabel('Признак')
plt.ylabel('Целевая переменная')
plt.legend()
plt.show()
```
