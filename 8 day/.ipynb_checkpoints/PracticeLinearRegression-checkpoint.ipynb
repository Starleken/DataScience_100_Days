```python
import pandas as pd
import sklearn as sk
from sklearn.linear_model import LinearRegression 
import matplotlib.pyplot as plt
import numpy as np
from sklearn.model_selection import train_test_split
    
```

```python
df = pd.read_csv('C:\\Users\\Starleken\\Desktop\\CarPrices.csv')
df = df[['horsepower', 'price', 'carwidth']].dropna()

X = df['horsepower'].to_numpy()
Y = df['price'].to_numpy()
X = X.reshape(-1,1)
Y = Y.reshape(-1,1)
model = LinearRegression()
model = model.fit(X,Y)

plt.scatter(X,Y)
plt.plot(X, model.predict(X), color='red')
plt.show()
```

```python
X = df['horsepower'].to_numpy()
Y = df['price'].to_numpy()
X = X.reshape(-1,1)
Y = Y.reshape(-1,1)

iteration_count = 100
alpha = 0.0001
bias = 0
slope = 0

count = len(X)
for i in range(iteration_count):
    predict = bias + X * slope
    
    bias_gradient = -np.sum(Y - predict) / count
    slope_gradient = -X.T.dot(Y - predict) / count
    bias = bias - bias_gradient * alpha
    slope = slope - slope_gradient * alpha
plt.scatter(X,Y)
plt.plot(X, bias + X * slope, color='red')
plt.show()

print(f'Изменение цены за единицу лошадиных сил: {bias + 1 * slope}')
```

```python
X = df['carwidth'].to_numpy()
Y = df['price'].to_numpy()

X = X.reshape(-1,1)
Y = Y.reshape(-1,1)

model = LinearRegression()
model = model.fit(X,Y)

plt.scatter(X,Y)
plt.plot(X, model.predict(X), color='red')
plt.show()
```
