```python
import pandas as pd
import sklearn as sk
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
```

```python
dataset = sk.datasets.make_blobs(centers=10)

print(len(dataset[0]))
print(len(dataset[1]))

plt.scatter(dataset[0][:,0], dataset[0][:,1])
plt.show()
```

```python
model = KMeans(n_clusters=10)
model.fit(dataset[0])
```

```python
plt.scatter(dataset[0][:,0], dataset[0][:,1],c=model.labels_)
plt.show()
```
